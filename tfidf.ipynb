```python
from konlpy.tag import Okt
twitter_tag=Okt()
print(twitter_tag.morphs(df.review[0]))
print(twitter_tag.nouns(df.review[0]))
print(twitter_tag.pos(df.review[0]))
```

```python
def new_tokenizer(doc):
    return [
        token
        for token, pos in twitter_tag.pos(doc)
        if pos in ['Noun','Verb','Adjective']
    ]
print(new_tokenizer(df.review[0]))
```

```python
from konlpy.tag import Okt
t=Okt()
tokens_const=[]
for idx, row in df.iterrows():
    #print(row['review'])
    print(t.morphs(row['review']))
    tokens_const.append(t.morphs(row['review']))
```

```python
print(type(tokens_const))
print(len(tokens_const))
print(type(tokens_const[0]))
print(len(tokens_const[0]))
print(tokens_const[0])
```

```python
len(tokens_const[0][-26])
```

```python
def longerthanone(x):
    return len(x)>1

def engorkr(x):
    if x.isalpha():
        return True
    return False
results = list(filter(longerthanone, tokens_const[0]))
print(type(results))
```

```python
print(engorkr("ekfm"))
print(engorkr("다른"))
print(engorkr(".."))
```

```python
tokens=[list(filter(longerthanone, token)) for token in tokens_const]
print(type(tokens))
print(len(tokens))
print(type(tokens[0]))
print(len(tokens[0]))
print(len(tokens[1]))
print(len(tokens[2]))
print(tokens[0])
print(tokens[1])
print(tokens[2])
```

```python
tokens=[list(filter(engorkr, token)) for token in tokens]
print(type(tokens))
print(len(tokens))
print(type(tokens[0]))
print(len(tokens[0]))
print(len(tokens[1]))
print(len(tokens[2]))
print(tokens[0])
print(tokens[1])
print(tokens[2])
```

```python
word_count={}
for text in tokens:
    for word in text:
        word_count[word]=word_count.get(word,0)+1
        
sorted_features=sorted(word_count, key=word_count.get, reverse=True)
print('num of features:', len(sorted_features))
for word in sorted_features[:10]:
    print(f"count of '{word}': {word_count[word]}", end=', ')
```

```python
word_features=sorted_features[:1000]
print(word_features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
start=len(df.review[0])//2
source=df.review[0][-start:]
source
```

```python
print(type(df['review']))
review_list=np.array(df['review'].tolist()).reshape(-1,1)
print(review_list.shape)
print(review_list)
```

```python
print(type(df['review']))
review_lists=df['review'].tolist()
print(review_lists[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(vocabulary=word_features)
reviews_tf=tf.fit_transform(review_lists)
source_tf=tf.transform([source])
sim_result_tf=cosine_similarity(source_tf,reviews_tf)

print('가장 유사한 리뷰 인덱스:', np.argmax(sim_result_tf[0]))
```

```python
print('#TF-IDF 벡터에 대해 가장 유사한 리뷰 순으로 인덱스:', (-sim_result_tf[0]).argsort()[:10])
```
